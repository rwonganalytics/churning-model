```python
dataset_for_features = pd.read_csv("../data/interim/churning_data_model.csv")
dataset_for_features = dataset_for_features.drop(
    dataset_for_features.columns[0], axis=1)
dataset_for_features.head()
```

#### Codificamos la Variable Gender y el Target

```python
dataset_for_features = pd.get_dummies(
    dataset_for_features,
    columns=['Attrition_Flag', 'Gender'],
    drop_first=True)

dataset_for_features = dataset_for_features.rename(
    columns={'Gender_M': 'Gender', 'Attrition_Flag_Existing Customer': 'Attrition_Flag'})

dataset_for_features['Gender'] = dataset_for_features['Gender'].astype(int)
dataset_for_features['Attrition_Flag'] = dataset_for_features['Attrition_Flag'].astype(int)

dataset_for_features.head()
```

#### Codificamos la Variable Card_Category con Frequency Encoding

```python
card_cat_encoder = dataset_for_features['Card_Category'].value_counts(
    normalize=False)

dataset_for_features['Card_Category'] = dataset_for_features['Card_Category'].map(
    card_cat_encoder)

dataset_for_features.head()
```
